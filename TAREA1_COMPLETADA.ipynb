#### Ejercicio:

+ Una barra rígida, uniforme, de masa total ${\text{m}}$ y longitud ${\text{L}_{\text{2}}}$, está suspendida en el punto ${\text{0}}$ por una cuerda de longitud ${\text{L}_{\text{1}}}$. Sobre la barra actúa una fuerza horizontal ${\text{F}}$ como se muestra en la Figura 1.



**Figura 1: Péndulo de una cuerda y una barra.**

+ Usar los desplazamientos angulares $\theta_{\text{1}}$ y $\theta_{\text{2}}$ para definir la posición, velocidad y aceleración del centro de masa $\text{C}$, en términos de los ejes del cuerpo $\hat{r}_{2}$ y $\hat{\theta}_{2}$

# Desarrollo:

Para comenzar a resolver el problema, primero se tomará en cuenta los vectores de posición en la siguiente Figura:



**Figura 2: Posición del centro de masa**

Se puede observar que el vector de posición de centro de masa es la suma de los vectores $\vec{r}_{A}$ y $\vec{r}_{AC}$, por lo tanto tenemos:

\begin{align*}
{\vec{r}_{C} = \vec{r}_{A} + \vec{r}_{AC}}
\end{align*}

Se puede definir en función de $\hat{u}_{r1}$ el vector $\vec{r}_{A}$: 

\begin{align*}
{\vec{r}}_{\text{A}}={\text{L}}_{\text{1}} \cdot {\hat{u}}_{\text{r1}}
\end{align*}

También se define en función de $\hat{u}_{r2}$ el vector $\vec{r}_{AC}$:

\begin{align*}
{\vec{r}}_{\text{AC}}= \frac{{\text{L}}_{\text{2}}}{2} \cdot {\hat{u}}_{\text{r2}}
\end{align*}

Reemplazando $\vec{r}_{AC}$ y $\vec{r}_{A}$ en $\vec{r}_{C}$:

\begin{align*}
{\vec{r}}_{\text{C}}={\text{L}}_{\text{1}} \cdot {\hat{u}}_{\text{r1}}+\frac{{\text{L}}_{\text{2}}}{2} \cdot {\hat{u}}_{\text{r2}}
\end{align*}



Análogamente, se tienen los vectores de velocidad:

\begin{align*}
{\vec{v}_{C} = \vec{v}_{A} + \vec{v}_{AC}}
\end{align*}

\begin{align*}
{\vec{v}}_{\text{A}}={\text{L}}_{\text{1}} \cdot {\dot{\theta}}_{\text{1}} \cdot {\hat{u}}_{{\theta}{\text1}}
\end{align*}

\begin{align*}
{\vec{v}}_{\text{A}}= \frac{{\text{L}}_{\text{2}}}{2} \cdot {\dot{\theta}}_{\text{2}} \cdot {\hat{u}}_{{\theta}{\text2}}
\end{align*}


\begin{align*}
{\vec{v}}_{\text{C}}= {\text{L}}_{\text{1}} \cdot {\dot{\theta}}_{\text{1}} \cdot {\hat{u}}_{{\theta}{\text1}} + \frac{{\text{L}}_{\text{2}}}{2} \cdot {\dot{\theta}}_{\text{2}} \cdot {\hat{u}}_{{\theta}{\text2}}
\end{align*}

Asumimos que el péndulo tiene un movimiento y usamos la siguiente formula:

\begin{align*}
{\vec{v}}_{\text{C}}={\dot{\vec{r}}}_{\text{C}}=\vec{r} \cdot \dot{\theta} \cdot {\hat{\theta}}
\end{align*}

Así, para la aceleración, tenemos:

\begin{align*}
{\vec{a}}_{\text{C}}= - {\text{L}}_{\text{1}} \cdot {\dot{\theta}^2}_{\text{1}} \cdot {\hat{u}_{\text{r1}} + {\text{L}}_{\text{1}}} \cdot {\ddot{\theta}}_{\text{1}} \cdot {\hat{u}}_{\theta1} - \frac{{\text{L}}_{\text{2}}}{2} \cdot \dot{\theta}^{2}_{\text{2}} \cdot {\hat{u}}_{r2} + \frac{{\text{L}}_{\text{2}}}{2} \cdot \dot{\theta}^{2}_{\text{2}} \cdot {\hat{u}}_{\theta2}
\end{align*}

###### Como las ecuaciones de posición y de aceleración están en términos de vectores unitarios, se encuentran las siguientes relaciones:



**Figura 3: Posición del centro de masa con los elementos necesrios para calcular ${\hat{u}}_{\text {r1}}$**

A partir de la Figura 3 se llega a que ${\hat{u}}_{\text {r1}}$ se puede descomponer de la siguiente manera:

\begin{align*}
\hat{u}_{\text r1} = \cos(\theta_{\text 2}-\theta_{\text 1})\cdot\hat{u}_{\text r2} - \sin(\theta_{\text 2}-\theta_{\text 1})\cdot\hat{u}_{\theta2}
\end{align*}




Análogamente se puede hacer lo mismo con la siguiente Figura para ${\hat{u}}_{\theta 1}$



**Figura 4: Posición del centro de masa con los elementos necesrios para calcular ${\hat{u}}_{\theta 1}$**

\begin{align*}
{\hat{u}}_{\theta 1} = \sin(\theta_{\text 2}-\theta_{\text 1})\cdot\hat{u}_{\text r2} + \cos(\theta_{\text 2}-\theta_{\text 1})\cdot\hat{u}_{\theta2}
\end{align*}


###### Se reemplaza ${\hat{u}}_{\text {r1}}$ y ${\hat{u}}_{\theta 1}$ en las ecuacioneas de posición, velocidad y aceleración :

Para la ecuación de posición:


\begin{align*}
{\vec{r}}_{\text{C}}={\text{L}}_{\text{1}} \cdot [\cos(\theta_{\text 2}-\theta_{\text 1})\cdot\hat{u}_{\text r2} - \sin(\theta_{\text 2}-\theta_{\text 1})\cdot\hat{u}_{\theta2}] +\frac{{\text{L}}_{\text{2}}}{2} \cdot {\hat{u}}_{\text{r2}}
\end{align*}


Reorganizamos los términos y nos queda:


\begin{align*}
{\vec{r}}_{\text{C}}= [{\text{L}} _{\text{1}} \cdot \cos(\theta_{\text 2}-\theta_{\text 1} +\frac{{\text{L}}_{\text{2}}}{2}] \cdot \hat{u}_{\text r2} - {\text{L}}_{\text{1}} \sin(\theta_{\text 2}-\theta_{\text 1})\cdot\hat{u}_{\theta2} 
\end{align*}


Se continua con la ecuación de velocidad:


\begin{align*}
{\vec{v}}_{\text{C}}= {\text{L}}_{\text{1}} \cdot {\dot{\theta}}_{\text{1}} \cdot [\sin(\theta_{\text 2}-\theta_{\text 1})\cdot\hat{u}_{\text r2} + \cos(\theta_{\text 2}-\theta_{\text 1})\cdot\hat{u}_{\theta2}] + \frac{{\text{L}}_{\text{2}}}{2} \cdot {\dot{\theta}}_{\text{2}} \cdot {\hat{u}}_{{\theta}{\text2}}
\end{align*}


Reorganizamos los términos y nos queda:


\begin{align*}
{\vec{v}}_{\text{C}}= {\text{L}}_{\text{1}} \cdot {\dot{\theta}}_{\text{1}} \cdot \sin(\theta_{\text 2}-\theta_{\text 1})\cdot\hat{u}_{\text r2} + [{\text{L}}_{\text{1}} \cdot {\dot{\theta}}_{\text{1}} \cdot \cos(\theta_{\text 2}-\theta_{\text 1})+ \frac{{\text{L}}_{\text{2}}}{2} \cdot {\dot{\theta}}_{\text{2}}] \cdot\hat{u}_{\theta2} 
\end{align*}


Finalmente la ecuación de aceleración:

\begin{align*}
{\vec{a}}_{\text{C}}= - {\text{L}}_{\text{1}} \cdot {\dot{\theta}^2}_{\text{1}} \cdot [\cos(\theta_{\text 2}-\theta_{\text 1})\cdot\hat{u}_{\text r2} - \sin(\theta_{\text 2}-\theta_{\text 1})\cdot\hat{u}_{\theta2}] + {\text{L}}_{\text{1}} \cdot {\ddot{\theta}}_{\text{1}} \cdot[\sin(\theta_{\text 2}-\theta_{\text 1})\cdot\hat{u}_{\text r2} + \cos(\theta_{\text 2}-\theta_{\text 1})\cdot\hat{u}_{\theta2}] - \frac{{\text{L}}_{\text{2}}}{2} \cdot \dot{\theta}^{2}_{\text{2}} \cdot {\hat{u}}_{r2} + \frac{{\text{L}}_{\text{2}}}{2} \cdot \dot{\theta}^{2}_{\text{2}} \cdot {\hat{u}}_{\theta2}
\end{align*}

Reorganizamos los términos y nos queda:
\begin{align*}
{\vec{a}}_{\text{C}}= [{\text{L}}_{\text{1}} \cdot {\ddot{\theta}}_{\text{1}} \cdot \sin(\theta_{\text 2}-\theta_{\text 1}) - {\text{L}}_{\text{1}} \cdot {\dot{\theta}^2}_{\text{1}} \cdot \cos(\theta_{\text 2}-\theta_{\text 1})- \frac{{\text{L}}_{\text{2}}}{2} \cdot \dot{\theta}^{2}_{\text{2}}] \cdot {\hat{u}}_{r2} + [{\text{L}}_{\text{1}} \cdot {\ddot{\theta}}_{\text{1}} \cdot \cos(\theta_{\text 2}-\theta_{\text 1}) + {\text{L}}_{\text{1}} \cdot {\dot{\theta}^2}_{\text 1}\cdot sin(\theta_{\text 2}-\theta_{\text 1}) + \frac{{\text{L}}_{\text{2}}}{2} \cdot \ddot{\theta}_{\text{2}}] \cdot {\hat{u}}_{\theta2}
\end{align*}



**Se derivan las ecuaciones de movimiento**

Primero observamos el siguiente diagrama de cuerpo libre:



**Figura 5: Diagrama de cuerpo libre de la barra*

Se tiene que $\text{T}$ es la tensión de la cuerda y $\text{mg}$ el peso de la barra. Además las ecuaciones de fuerza en término de los componentes de los ejes del cuerpo:

\begin{align*}
{\text{F}_{r2}} = \text{m} \cdot \text{a}_{cr2}
\end{align*}

\begin{align*}
{\text{F}_{\theta2}} = \text{m} \cdot \text{a}_{\text{c}\theta2}
\end{align*}

También se tiene el momento dado por la siguiente ecuación:

\begin{align*}
{\text{M}_{c}} = \text{I}_{c} \alpha = \text{I}_{c} {\ddot{\theta}}_{\text{2}}
\end{align*}

Donde: 

$\text{M}_{c}$ es el momento con respecto al centro de masa,  $\text{I}_{c}$ es el momento de inercia y $\alpha$ la aceleración angular. 

Esta ecuación se la puede desglosar de la siguiente manera, tomando en cuenta el diagrama de cuerpo libre de la Figura 5:

\begin{align*}
{\text{M}_{c}} = \text{M}_{F} + \text{M}_{T} + \text{M}_{mg}
\end{align*}

\begin{align*}
{\text{M}_{F}} = \text{r}_{f} \times \text{F}_{f}
\end{align*}
\begin{align*}
{\text{M}_{F}} = \frac{\text{L}_{2}}{2} \cdot \text{F} \cos(\theta_2)
\end{align*}


\begin{align*}
{\text{M}_{T}} = \text{r}_{T} \times \text{F}_{T}
\end{align*}
\begin{align*}
{\text{M}_{T}} = \frac{\text{L}_{2}}{2} \cdot -\text{T} \sin(\theta_2 - \theta_1)
\end{align*}

\begin{align*}
{\text{M}_{mg}} = \text{r}_{mg} \times \text{F}_{mg}
\end{align*}

\begin{align*}
 {\text{r}_{mg}} = 0 ; \text{por lo tanto:}
\end{align*}

\begin{align*}
{\text{M}_{mg}} = 0
\end{align*}

Asi finalmente tenemos:

\begin{align*}
{\text{M}_{c}} = \frac{\text{L}_{2}}{2} \cdot \text{F} \cos(\theta_2) - \frac{\text{L}_{2}}{2} \cdot \text{T} \sin(\theta_2 - \theta_1)
\end{align*}

Para las ecuaciones de fuerza, se puede llegar a las siguientes ecuaciones a partir de la Figura 5:

\begin{align*}
{\text{F}_{r2}} = \text{F} \sin(\theta_2) + \text{mg} \cos(\theta_2) -\text{T} \sin(\theta_2 - \theta_1)
\end{align*}


\begin{align*}
{\text{F}_{\theta2}} = \text{F} \cos(\theta_2) - \text{mg} \sin(\theta_2) + \text{T} \sin(\theta_2 - \theta_1)
\end{align*}



Por otro lado, se sabe que $\text{a}_{cr2}$ y $\text{a}_{\theta r2}$ son simplemente los coeficientes de ${\hat{u}}_{\theta2}$ y ${\hat{u}}_{\theta2}$. Así reemplazando en las fórmulas obtenidas se tiene:
\begin{align*}
\text{F} \sin(\theta_2) + \text{mg} \cos(\theta_2) - \text{T} \sin(\theta_2 - \theta_1)= \text{m} [{\text{L}}_{\text{1}} \cdot {\ddot{\theta}}_{\text{1}} \cdot \sin(\theta_{\text 2}-\theta_{\text 1}) - {\text{L}}_{\text{1}} \cdot {\dot{\theta}^2}_{\text{1}} \cdot \cos(\theta_{\text 2}-\theta_{\text 1})- \frac{{\text{L}}_{\text{2}}}{2} \cdot \dot{\theta}^{2}_{\text{2}}]
\end{align*}

\begin{align*}
\text{F} \cos(\theta_2) - \text{mg} \sin(\theta_2) + \text{T} \sin(\theta_2 - \theta_1)= \text{m} [{\text{L}}_{\text{1}} \cdot {\ddot{\theta}}_{\text{1}} \cdot \cos(\theta_{\text 2}-\theta_{\text 1}) + {\text{L}}_{\text{1}} \cdot {\dot{\theta}^2}_{\text 1}\cdot sin(\theta_{\text 2}-\theta_{\text 1}) + \frac{{\text{L}}_{\text{2}}}{2} \cdot \ddot{\theta}_{\text{2}}]
\end{align*}


Ahora para el Momento tenemos:

\begin{align*}
\frac{\text{L}_{2}}{2} \cdot \text{F} \cos(\theta_2) - \frac{\text{L}_{2}}{2} \cdot \text{T} \sin(\theta_2 - \theta_1)= \text{I}_{c} {\ddot{\theta}}_{\text{2}}
\end{align*}

$\text{I}_{c}$ puede ser expresado de la siguiente manera:
\begin{align*}
\text{I}_{c}=\int_{}^{} x^2 \,du
\end{align*}

\begin{align*}
\text{du}=\frac{m}{L} \cdot text{dx}
\end{align*}

\begin{align*}
\text{I}_{c}=\frac{m}{L} \cdot \int_{\frac{\text{-L}}{2}}^{\frac{{L}}{2}} x^2 \,dx
\end{align*} 

\begin{align*}
\text{I}_{c} =\frac{m}{L} \cdot [\frac{\frac{{\text{L}}^3}{2}}{3} + \frac{\frac{{\text{-L}}^3}{2}}{3}]
\end{align*} 

\begin{align*}
\text{I}_{c} =\frac{m}{L} \cdot [\frac{{\text{L}}^3}{24} + \frac{{\text{L}}^3}{24} ]
\end{align*} 

\begin{align*}
\text{I}_{c} =\frac{m \cdot {\text{L}}^2}{12} 
\end{align*} 

Reemplazando en la ecuación de momento, queda finalmente:

\begin{align*}
\frac{\text{L}_{2}}{2} \cdot \text{F} \cos(\theta_2) - \frac{\text{L}_{2}}{2} \cdot \text{T} \sin(\theta_2 - \theta_1)= \frac{m \cdot {\text{L}}^2}{12} {\ddot{\theta}}_{\text{2}}
\end{align*}

Se puede ver que el movimiento angular de la barra está totalmente definido por los movimientos angulares $\theta_1$ y $\theta_2$, y tenemos tres ecuaciones de movimiento. el tercer término que no se conoce es la Tensión ($\text{T}$), la cual si no es de interés, puede ser eliminada de la ecuación final y reducir las ecuaciones a dos. Esto se puede hacer por ejemplo al igualar una de las ecuaciones finales de fuerza a ($\text{T}$) y reemplazar en en las otras dos ecuaciones, así quedando con ds funciones que dependen simplemente de $\theta_1$ y $\theta_2$.
