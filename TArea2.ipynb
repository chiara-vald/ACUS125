# Resumen: Sistema vibratorio con amortiguamiento viscoso de un grado de libertad

## Introducción

+ Un sistema que vibra se lo puede tomar como una composición de partes que almacenan energía. En el caso de un resorte está la energía potencial y en el caso de una masa la cinética. También cuenta con una parte que disipa energía (amortiguador).

 + Existen ciertos parámetros discretos de un sistema mecánico que vibra, también denominado sistema de cuerpo rígido:
   + 1) La masa: 
Accionada por una fuerza $F_m$ , proporcional a aceleración de la masa. Esta componente almacena energía cinética cuando la velocidad aumenta y viceversa.
   + 2) Amortiguamiento: 
Accionado por una fuerza $F_d$ , proporcional a la velocidad de la masa. Esta componente se caracteriza por hacer que el sistema disipe o pierda energía.
   + 3) Elasticidad: 
Accionado por una fuerza $F_s$ , proporcional a la posición de la masa. Se caracteriza por almacenar energía potencial.

+ Denominar discretos a los parámetros es suponer que el sistema es una combinación de masas interconectadas por resortes flexibles y elementos amortiguadores.


### GRADOS DE LIBERTAD

+ Son el número de coordenadas independientes que se necesitan para determinar completamente la posición de cada parte de un sistema vibratorio.
+ Un ejemplo de un sistema de un grado de libertad es el caso de un péndulo simple, donde la posición del péndulo depende completamente del desplazamiento angular $\theta({\text{t}})$.

### RESORTES

+ Es el parámetro que une las partículas que vibran en un cuerpo. La masa del resorte es tan pequeña en comparación a la de la masa del sistema que se la desprecia.

+ Comúnmente, la fuerza del resorte es una función que depende de la posición debido a su estiramiento $F=f(x)$ (donde $x$ puede ser + o - ).
+ Todos los resortes son "no lineales". Sin embargo hay muchos casos donde la extensión $x$ se puede asumir suficientemente pequeña como para que los términos no lineales de su respectiva ecuación se desprecien teniendo como resultado final: 

\begin{align}
{\text{F}}&=-{\text{k}}{\text{x}} 
\label{eq1}
\end{align}

Donde $k$ es la constante del resorte.



Con esto tenemos  que el amortiguador ayuda a que la energía del sistema se disipe, haciendo que su amplitud decaiga exponencialmente, tanto la amplitud de vibración libre como forzada.
Al agregar este término tenemos, al replantear la ecuación diferencial de movimiento:

\begin{align}
m\ddot{x}(t)= -c\dot{x}(t)-k,x(t)
\end{align}

\begin{align}
m\ddot{x}(t)&+c\dot{x}(t)+k,x(t)=0
\end{align}
Siendo:
\begin{align}
m\ddot{x}(t) = Fuerzas \ inerciales
\end{align}

\begin{align}
c\dot{x}(t) = Fuerzas \ disipativas
\end{align}

\begin{align}
kx(t) = Fuerzas \ elásticas
\end{align}

## Vibración libre de un sistema lineal




Utilizando la segunda ley de Newton en un sistema masa-resorte linal tenemos que:
\begin{align}
\sum_{}^{}F&=m\ddot{x}(t)
\end{align}
\begin{align}
m\ddot{x}(t)&=-kx(t)
\end{align}

Puesto que el sistema masa-resorte en la práctica no vibra indefinidamente, añadimos una componente al módelo matemático, la fuerza de amortiguamiento, la cual es proporcional a la velocidad. Este término lineal ($c\dot{x}(t)$) representa el decaimiento y el contacto con el fluído. Se toma todo este término como amortiguamiento viscoso.

Con esto se tiene que el amortiguador ayuda a que la energía del sistema se disipe, haciendo que su amplitud decaiga exponencialmente, tanto la vibración libre como forzada.
Se agrega este término a la ecuación diferencial de movimiento:

\begin{align}
m\ddot{x}(t)&=-c\dot{x}(t)-kx(t)
\end{align}

\begin{align}
m\ddot{x}(t)&+c\dot{x}(t)+kx(t)=0
\end{align}


#### Parámetros necesarios:

\begin{align*}
\zeta &=\frac{c}{2\,\sqrt{k\,m}},\quad {\text{razón de amortiguamiento (cantidad que no tiene unidades)}}
\label{eq4}
\end{align*}

\begin{align*}
\omega_{0}&=\sqrt{\frac{k}{m}},\quad {\text{frecuencia angular natural de movimiento}}\,\bigg( \frac{rad}{s}\bigg)
\label{eq3}
\end{align*}


## Casos de amortiguamiento:


La ecuación característica está dada por:

\begin{align*}
\alpha&= \omega_{0}\big(-\zeta\pm\sqrt{\zeta^{2}-1}\big)
\label{eq11}
\end{align*}


Se tienen las siguietes condiciones iniciales:

\begin{align*}
x(0)&=x_{0}
%\label{eq24}\tag{24}
\end{align*} 
\begin{align*}
\dot{x}(0)&=\dot{x}_{0}
%\label{eq25}\tag{25}
\end{align*}

### Caso1: Vibración libre no amortiguada


En este caso: $\zeta=0$, obteniendo como resultado:

\begin{align*}
\alpha&= \pm\, j\omega_{0}
\end{align*}


Suponiendo que la solución es del tipo :

\begin{align*}
x(t)&=B_{1}e^{j\omega_{0}t}+B_{2}e^{-j\omega_{0}t}
\end{align*}

Con $B_1$ y $B_2$ siendo constantes de integración.

Utilizando la propiedad:

\begin{align*}
Ae^{ \pm\,jx}=A(cos(x) \pm\, jsen(x))
\end{align*}

Aplicando esta misma en la solución que obtuvimos anteriormente nos queda:

\begin{align*}
x(t)&=B_{1}(cos(\omega_{0}t) + jsen(\omega_{0}t))+B_{2}(cos(\omega_{0}t) - jsen(\omega_{0}t)
\end{align*}

Factorizando llegamos al resultado:

\begin{align*}
x(t)&=(B_{1}+B_{2})cos(\omega_{0}t) + j(B_{1}-B_{2})sen(\omega_{0}t)
\end{align*}

\begin{align*}
x(t)&=C_{1}cos(\omega_{0}t) + C_{2}sen(\omega_{0}t)
\end{align*}

Donde $C_{1}=(B_{1}+B_{2})$ y $C_{2}=j(B_{1}-B_{2})$

Derivamos la función con respecto al tiempo:

\begin{align*}
\dot{x}(t)&=-\omega_{0}C_{1}sen(\omega_{0}t) + \omega_{0}C_{2}cos(\omega_{0}t)
\end{align*}

De esta forma tenemos dos funciones con dos constantes ($C_{1}$ y $C_{2}$) y para obtenerlas utilizaremos las condiciones iniciales:

De esta forma al reemplazar las condiciones en las ecuaciones tenemos que los valores correspondientes de $C_{1}$ y $C_{2}$ son:

\begin{align*}
C_{1}=x_{0}
\end{align*}


\begin{align*}
C_{2}=\frac{\dot{x}}{\omega_{0}}
\end{align*}

Para finalizar reemplazandolo en la ecuación:

\begin{align*}
x(t)&=x_{0}cos(\omega_{0}t) + \frac{\dot{x}}{\omega_{0}}sen(\omega_{0}t)
\end{align*}


```python
#Gráfico de vibración libre no amortiguada

t     = np.linspace(0,15,2000)
c1    = 100 # [kg/s]
c2    = 200 # [kg/s]
k     = 9000 # [N/m]
m     = 200 # [kg]
x0    = 0.5 # [m]
xdot_0 = 100 # [m/s]
xdot_01 = 40 # [m/s]

zeta_1 = c1/(2*np.sqrt(k*m))
zeta_2 = c2/(2*np.sqrt(k*m))
wn     = np.sqrt(k/m)


x1 = [x0*np.cos(wn*t)+(xdot_0/wn)*np.sin(wn*t)for t in t]
x2 = [x0*np.cos(wn*t)+(xdot_01/wn)*np.sin(wn*t)for t in t]


plt.figure(figsize=(15,7))
plt.plot(t, x1, color='b', label='$c_1$= 100 [kg/s]')
plt.plot(t, x2, color='r', label='$ c_2$= 100 [kg/s]')
plt.xlabel('tiempo [s]',fontsize=16)
plt.ylabel('$x(t)[m]$ ',fontsize=16)
plt.title('Ejemplos de vibraciones no amortiguados', fontsize=20)
plt.legend(loc=4);
plt.grid()
plt.plot(t,x1)
plt.plot(t,x2)
```

### Caso 2: Vibración libre sub-amortiguada

En este caso: $0<\zeta<1$, por lo tanto la ecuación característica $\alpha$, atenua exponencialmente la amplitud y será un complejo conjugado.

\begin{align*}
\alpha&= \omega_{0}\big(-\zeta\,\pm\,j\,\sqrt{1-\zeta^{2}} \big)
\end{align*}

Teniendo como solución general:

\begin{align*}
x(t)&=B_{1}e^{(-\zeta\omega_{0}-j\omega_{0}\sqrt{1-\zeta^{2}})t}+B_{2}e^{(-\zeta\omega_{0}+j\omega_{0}\sqrt{1-\zeta^{2}})t}
\end{align*}

Utilizando la propiedad mostrada con anterioridad obtenemos el siguiente resultado:

\begin{align*}
x(t)&=e^{-\zeta\omega_{0}t}[B_{1}cos(\sqrt{1-\zeta^2}\omega_{0}t) - jB_{1}sen(\sqrt{1-\zeta^2}\omega_{0}t) + B_{2}cos(\sqrt{1-\zeta^2}\omega_{0}t) + jB_{2}sen(\sqrt{1-\zeta^2}\omega_{0}t)]
\end{align*}

Reemplazando en los el resultado tenemos:

\begin{align*}
x(t)&=e^{-\zeta\omega_{0}t}[C_{1}cos(\sqrt{1-\zeta^2}\omega_{0}t) + jC_{2}sen(\sqrt{1-\zeta^2}\omega_{0}t)]
\end{align*}

Donde $C_{1}=(B_{1}+B_{2})$ y $C_{2}=j(B_{2}-B_{1})$

Si tomamos las condiciones iniciales anteriores, nos quedará finalmente:

\begin{align*}
C_{1}=x_{0}
\end{align*}

\begin{align*}
C_{2}=\frac{\dot{x}+\zeta\omega_{0}x_{0}}{\omega_{0}\sqrt{1-\zeta^2}}
\end{align*}

Finalmente reemplazamos los valores $C_{1}$ y $C_{2}$:

\begin{align*}
x(t)&=e^{-\zeta\omega_{0}t}[x_{0}cos(\sqrt{1-\zeta^2}\omega_{0}t) + j\frac{\dot{x}+\zeta\omega_{0}x_{0}}{\omega_{0}\sqrt{1-\zeta^2}}sen(\sqrt{1-\zeta^2}\omega_{0}t)]
\end{align*}

Otra manera de solucionarlo es la siguiente:

La solución general para este problema:

\begin{align*}
x(t)&=B_{1}e^{(-\zeta\omega_{0}-j\omega_{d})\cdot t}+B_{2}e^{(-\zeta\omega_{0}+j\omega_{d})\cdot t}
\end{align*}

Donde: 

\begin{align*}
\omega_{d}&=\omega_{0}\sqrt{1-\zeta^{2}} \ \ \ \ \ \ \ \text[rad/s]
\end{align*}


Tambien se puede escribir el resultado de la siguiente manera:


\begin{align*}
x(t)&=Ae^{-\zeta\omega_{0}t}\,{\text{sen}}\big( \omega_{d}t-\phi_{d}\big)
\end{align*}


Donde, aplicando las condiciones de contorno A y $\phi$ son:

\begin{align*}
A&=\sqrt{x_{0}^{2}+\big(\frac{\dot{x}(0)+\zeta\omega_{0}x_{0}}{\omega_{d}}\big)^{2}}
\end{align*}


\begin{align*}
\omega_{d}&=\omega_{0}\sqrt{1-\zeta^{2}}
\end{align*}





```python
#Gráfico de Vibración libre sub-amortiguada usando dos amortiguamientos distintos

t     = np.linspace(0,15,2000)
c1    = 100
c2    = 200
k     = 9000
m     = 200
x0    = 0.5 
xdot_0 = 60

zeta_1 = c1/(2*np.sqrt(k*m))
zeta_2 = c2/(2*np.sqrt(k*m))
w0     = np.sqrt(k/m)
wd_1   = w0* np.sqrt(1- zeta_1**2)
wd_2   = w0* np.sqrt(1- zeta_2**2)
A1     = np.sqrt(x0**2+ ((xdot_0+zeta_1*w0*x0)/wd_1)**2)
A2     = np.sqrt(x0**2+ ((xdot_0+zeta_1*w0*x0)/wd_2)**2)
phid_1  = np.arctan((x0*wd_1)/(xdot_0+zeta_1*w0*x0))
phid_2  = np.arctan((x0*wd_2)/(xdot_0+zeta_2*w0*x0))

xt_1    = A1 * np.exp(-zeta_1*w0*t)*np.sin(wd_1*t-phid_1)
xt_2    = A2 * np.exp(-zeta_2*w0*t)*np.sin(wd_2*t-phid_2)





plt.figure(figsize=(25,15))
plt.plot(t, xt_1, color='c', label='$c_1$= 100 [kg/s]')
plt.plot(t, xt_2, color='r', label='$ c_2$= 100 [kg/s]')
plt.xlabel('tiempo (s)',fontsize=16)
plt.ylabel('$x(t)[m]$ ',fontsize=16)
plt.title('Ejemplos de vibraciones sub-amotiguadas', fontsize=20)
plt.legend(loc=4);
plt.grid()
plt.plot(t,xt_1)
plt.plot(t,xt_2)

```

### Caso 3: Vibrabración críticamente amortiguada

En este caso: $\zeta=1$ y la ecuación característica $\alpha$ tiene una sola raiz, por lo tanto el resultado es el siguiente:

\begin{align*}
\alpha&= -\omega_{0}
\end{align*}

Existen dos soluciones a la ecuación diferencial. Una es $e^{-\omega_{0}t}$ y la segunda es la misma pero multiplicada por $\text{t}$. De esta forma se obtiene la siguiente ecuación:

\begin{align*}
x(t)&=e^{-\omega_{0}\,t}\,(C_{1}+C_{2}\,t)
\end{align*}

Utilizando las condiciones iniciales:
\begin{align*}
C_{1}=x_{0}
\end{align*}

\begin{align*}
C_{1}=x_{0}
\end{align*}

Aplicando las condiciones iniciales se tiene:

\begin{align*}
x(t)&=e^{-\omega_{0}\,t}\,\big[\,x_{0}+\big(\dot{x}_{0}+\omega_{0}x_{0}\big)\cdot t\,\big]
\end{align*}

```python
#Gráfico de Vibración libre criticamente amortiguada

t     = np.linspace(0,2,200)

w_0   = 5 
m     = 300
x_0   = 30
x_dot0= 600

x_t   = np.exp(-w_0*t)*(x_0+(x_dot0 + w_0*x_0)*t)

plt.figure(figsize=(25,15))

plt.plot(t, x_t, color='c', label= '')

plt.xlabel('tiempo [s]',fontsize=16)
plt.ylabel('$x(t)[m]$ ',fontsize=16)
plt.title('Ejemplo de vibración criticamente amortiguada', fontsize=20)
plt.grid()
plt.plot(t,x_t)
```

### Caso4: Vibración sobreamortiguada

En este caso: $\zeta>1$

\begin{align*}
\alpha&= \omega_{0}\big(-\zeta\,\pm\,\sqrt{\zeta^{2}-1} \big)
\end{align*}

La solución es la siguiente:

\begin{align*}
x(t)&=e^{-\omega_{0}t}\bigg[C_{1}e^{-\omega_{0}\big(\sqrt{\zeta^{2}-1}\big)\,t}+C_{2}e^{\omega_{0}\big(\sqrt{\zeta^{2}-1}\big)\,t}\bigg]
\end{align*}


Aplicando las condiciones iniciales se obtiene:

\begin{align*}
x(t)&=\frac{e^{-\omega_{0}t}}{2\big(\sqrt{\zeta^{2}-1}\big)}\,\cdot\bigg\{\,\bigg[\frac{\dot{x}_{0}}{\omega_{0}}+x_{0}\,\big(\zeta+\sqrt{\zeta^{2}-1}\,\big)\bigg]\,e^{\omega_{0}\big(\sqrt{\zeta^{2}-1}\big)\,t}\,+\,\bigg[-\frac{\dot{x}_{0}}{\omega_{0}}+x_{0}\,\big(-\zeta+\sqrt{\zeta^{2}-1}\,\big)\bigg]\,e^{-\omega_{0}\big(\sqrt{\zeta^{2}-1}\big)\,t}\,\bigg\}
\end{align*}


```python
#Gráfico de Vibración libre sobreamortiguada 

w0   = 3
zeta = 1.2
x0   = 1 
v1   = 9

x  = (np.exp(-zeta*w0*t)/2*np.sqrt(zeta**2-1))*((v1/w0+x0*(zeta+np.sqrt(zeta**2-1)))*np.exp(w0*t*np.sqrt(zeta**2-1))+(-v1/w0+x0*(-zeta+np.sqrt(zeta**2-1)))*np.exp(-w0*t*np.sqrt(zeta**2-1)))

plt.figure(figsize=(25,15))

plt.plot(t, x, color='b', label= '')

plt.xlabel('tiempo [s]',fontsize=16)
plt.ylabel('$x(t)[m]$ ',fontsize=16)
plt.title('Ejemplo de vibración sobreamortiguada', fontsize=20)
plt.grid()
plt.plot(t,x)
```
